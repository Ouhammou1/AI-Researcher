```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exemple y = 2 * x
x = 1 
y = 2 * x
print(f"When x = {x}, y = {y}")

x = 5 
y = 2 * 3
print(f"When x = {x}, y = {y}")

x_values = np.linspace(0,10 ,100)
print(x_values)
print(f"lenght of x_values is {len(x_values)}")
print(type(x_values))
```

```python
y_values = 2 * x_values
print(y_values)
print(len(y_values))
```

```python

plt.figure(figsize=(10,7))
plt.plot(x_values , y_values , 'y*' , linewidth=2 , label='y=2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function y =2x')
plt.grid(True)
plt.legend()
plt.show()
```

```python

k_values = [1, 2, 3, 0.5, -1]
x = np.linspace(-5,5,100)


plt.figure(figsize=(10,7))
for i in k_values:
    y = i * x
    plt.plot(x , y  , linewidth=2 , label=f'y={i} * x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function y =2x')
plt.grid(True)
plt.legend()
plt.show()
```

```python


x = np.linspace(-100,100,100)
print(len(x))
y = x**2

for i in range(0,10):
    print(f"When x ={x[i]} , y = {x[i]**2}")
plt.figure(figsize=(10,7))
plt.plot(x , y  , linewidth=2 , label=f'y={i} * x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function y =2x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x_cubic = np.linspace(-5, 5, 100)
y_cubic  = x_cubic **3

plt.figure(figsize=(7,4))
plt.plot(x_cubic , y_cubic , linewidth=3 , label=f'y=x **3')
plt.xlabel('x_cubic')
plt.ylabel('y_cubic')
plt.title("Cubic Function :y = x**3")
plt.grid()
plt.legend()
plt.show()
```

```python
x_sqrt = np.linspace(1 ,15, 100)
y_sqrt = np.sqrt(x_sqrt)

plt.figure(figsize=(7,4))
plt.plot(x_sqrt , y_sqrt , 'b-', linewidth=2 , label='sqrt(x)')
plt.xlabel('x_sqrt')
plt.xlabel('y_sqrt')
plt.title("Square function ")
plt.grid()
plt.legend()
plt.show()
```

```python
x_exp = np.linspace(-5,5,100)
y_exp = 2**x_exp

print(f'x = {x_exp[0]}')
print(f'x = {y_exp[0]}')


plt.figure(figsize=(7,5))
plt.plot(x_exp , y_exp , 'b-', linewidth=3 , label='2**x')
plt.show()
```

```python

x_sin = np.linspace(0,15*np.pi,100)
y_sin = np.sin(x_sin)


plt.figure(figsize=(7,5))
plt.plot(x_sin , y_sin ,'b-' ,label='y = sin(x)')
plt.show()
```

```python
x_sin = np.linspace(0,4*np.pi,100)
y_sin = np.cos(x_sin)


plt.figure(figsize=(7,5))
plt.plot(x_sin , y_sin ,'b-' ,label='y = sin(x)')
plt.show()
```
